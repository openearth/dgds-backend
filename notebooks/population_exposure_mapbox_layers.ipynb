```python
import geopandas as gpd
import pandas as pd
import numpy as np
```

```python
gdf = gpd.read_file(r'c:/Users/twigt_d/Downloads/_temp/mapbox/IPCC-WGI-reference-regions-v4.geojson')

gdf
```

```python
df = pd.read_csv(r'c:/Users/twigt_d/Downloads/_temp/mapbox/Population_exposed.csv')

df
```

```python
# Read mapping Region and index in zarr file from csv

df_zarr = pd.read_csv('..\zarr_index.csv')
print(df_zarr)
```

```python
thresholds = np.array([0.01, 0.1, 1, 5, 10, 25]) * 1e6

query_list = ['SLR == "SSP1-2.6" and Confidence =="Low" and Percentile == 50 and Population == "SSP1"',
              'SLR == "SSP1-2.6" and Confidence =="Medium" and Percentile == 50 and Population == "SSP1"',
             'SLR == "SSP2-4.5" and Confidence =="Medium" and Percentile == 50 and Population == "SSP2"',
             'SLR == "SSP5-8.5" and Confidence =="Low" and Percentile == 50 and Population == "SSP5"',
             'SLR == "SSP5-8.5" and Confidence =="Medium" and Percentile == 50 and Population == "SSP5"',
             'SLR == "SSP1-2.6" and Confidence =="Low" and Percentile == 50 and Population == "Present"',
             'SLR == "SSP1-2.6" and Confidence =="Medium" and Percentile == 50 and Population == "Present"',
             'SLR == "SSP2-4.5" and Confidence =="Medium" and Percentile == 50 and Population == "Present"',
             'SLR == "SSP5-8.5" and Confidence =="Low" and Percentile == 50 and Population == "Present"',
             'SLR == "SSP5-8.5" and Confidence =="Medium" and Percentile == 50 and Population == "Present"']

output_list = ['SSP1-26-Low_SSP',
               'SSP1-26-Medium_SSP',
               'SSP2-45-Medium_SSP',
               'SSP5-85-Low_SSP',
               'SSP5-85-Medium_SSP',
               'SSP1-26-Low_present',
               'SSP1-26-Medium_present',
               'SSP2-45-Medium_present',
               'SSP5-85-Low_present',
               'SSP5-85-Medium_present']

for i, query in enumerate(query_list):
    print(output_list[i])
    print(query)

    selected = df.query(query)
    rows = []
    for threshold in thresholds:
        for region, df_i in selected[selected.Pop_exposed > threshold].groupby('Region'):
            result = df_i.sort_values('Pop_exposed').head(n=1).copy()
            result['threshold'] = threshold             

            region_string = result['Region'].astype(str).values.tolist()

            for zarr_index, zarr_region in enumerate(df_zarr['Regions']):
                if str(region_string[0]) == zarr_region:
                    result['zarrIndex'] = zarr_index

            # Remove Greenland from list
            if str(region_string[0]) != "GIC":
                rows.append(result)

    pop_exposed = pd.concat(rows)

    pop_exposed
    # print(pop_exposed)
    
    pop_exposed_gdf = gdf.merge(pop_exposed, left_on='Acronym', right_on='Region')
    geometry = pop_exposed_gdf.geometry.apply(lambda geom: geom.centroid)
    # replace geometry with centers
    pop_exposed_gdf.geometry = geometry
    outname = "c:/Users/twigt_d/Downloads/_temp/mapbox/pop_exposed_" + output_list[i] + ".geojson"
    print(outname)
    pop_exposed_gdf.to_file(outname, driver='GeoJSON')
```
